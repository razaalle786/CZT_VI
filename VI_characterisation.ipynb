```python
import matplotlib.pyplot as plt
import pandas as pd

# Load data from CSV file
data = pd.read_csv('-5 V to 5 V.csv')

# Extract voltage and current data
voltage = data['Smu1_V(1)(1)']
current = data['Smu1_I(1)(1)']

# Create scatter plot
fig = plt.figure()
plt.scatter(voltage, current)

# Label the axes
plt.xlabel('Voltage (V)')
plt.ylabel('Current (A)')

# Show the plot
# plt.show()
plt.savefig('-5V_to_5V.png')
```
